(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
# Your code here

import random

random_numbers = [random.randint(1, 100) for i in range(100)]

print(f"Random list in ascending order: {sorted(random_numbers)}\n")

print(f"Random list in descending order: {sorted(random_numbers, reverse=True)}\n")

print(f"Average = {sum(random_numbers)/len(random_numbers)}")
```

# Question 2
(25 points). Write a program that will do the following string formatting tasks:

(1) Ask the user to enter two sentences.

(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

(3) Compare the two sentences. Print out a list of words that occur in both sentences.

(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc

```python
# Your code here
import string

input1 = input("Enter sentence 1: ") # "Hey!, How are you?"
input2 = input("Enter sentence 2: ") # "I'm Karthik!"
print()

# Code for separating words in a sentence with punctuation
def split_sentence(sentence):
    """This function splits the sentence into words and return a list"""
    result = [] 
    word = '' 

    for char in sentence:
        if char not in string.whitespace:
            if char not in string.ascii_letters + "'":
                if word: 
                        result.append(word)
                result.append(char)
                word = ''
            else:
                word = ''.join([word,char])

        else:
            if word:
                result.append(word)
                word = ''
    return result

list1 = split_sentence(input1)
list2 = split_sentence(input2)

print(f"Splitting input1: {list1}")
print(f"Splitting input2: {list2}")
print()

def compare_words_of_sentence(list1, list2):
    """This function compares two list of words and return common words of list."""
    result = []
    for word in list1:
        if word in list2:
            result.append(word)
            
    return result

print(f"Common words in the two lists: {compare_words_of_sentence(list1, list2)}")
print()


def remove_punctuation(list_of_words):
    """This function removes punctuation marks present in a given list of words and return a list."""
    result = []
    for word in list_of_words:
        if word.strip(string.punctuation):
            result.append(word.strip(string.punctuation))
    return result

list1 = remove_punctuation(list1)
list2 = remove_punctuation(list2)

print(f"Sentence 1 without punctuation: {list1}")
print(f"Sentence 2 without punctuation: {list2}")
print()

def count_vowels(sentence):
    """
    This function counts the number vowels present in a given string and return a dictionary.
    
    Input:
        sentence(str, mandatory) --> Pass the string you want to count the vowels.
    Output: (dictionary)
        return a dictionary with count of vowels.
    """
    
    dic = {'a': 0, 'e': 0, 'i':0, 'o':0, 'u': 0}
    for char in sentence.lower():
        if char in 'aeiou':
            if char in dic.keys():
                dic[char] += 1
            else:
                dic[char] = 1
    return dic

print(f"Vowels count in sentence 1: {count_vowels(input1)}")
print(f"Vowels count in sentence 2: {count_vowels(input2)}")
            
```

# Question 3 

(15 points). The formula that describes the volume V of a sphere with radius r is the following:

$ V=\frac{4}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

	r	V
	1	…
	2	…
	3	…


```python
# Your code here
import math 

def sphere_volume(radius):
    """
    This function calculates the volume of a sphere when a radius is passed.
    """
    return ((4/3) * math.pi * pow(radius,3))

print("r", "V", sep="\t")

for radius in range(1, 10+1):
    print(radius,sphere_volume(radius), sep='\t')
```

#### (1) (10 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

```python
print('Difference in the number of Occupied Housing Units from year to year:')
df_new_index = df.set_index('year')
for year in range(2008, 2014):
    difference = df_new_index[df_new_index['title_field']=='Occupied Housing Units']['value'][year] - \
        df_new_index[df_new_index['title_field']=='Occupied Housing Units']['value'][year+1]
    print(f"{year} - {year+1} = {difference}")
```

    2011 ----- -----

```python
print('year', 'Vacant Housing Units', 'Occupied Housing Units', sep='\t')

for year in range(2008, 2015):
    vacant_percent = ((df_new_index[df_new_index['title_field']=='Vacant Housing Units']['value'][year]) / (df_new_index[df_new_index['title_field']=='Total Housing Units']['value'][year]))*100
    occupied_percent = ((df_new_index[df_new_index['title_field']=='Occupied Housing Units']['value'][year]) / (df_new_index[df_new_index['title_field']=='Total Housing Units']['value'][year] ))*100

    print(year, str(vacant_percent)+'%', str(occupied_percent)+'%', sep='\t')
```
