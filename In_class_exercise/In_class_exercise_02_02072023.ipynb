```python
import requests
from bs4 import BeautifulSoup

# Make a request to the website
url = 'https://www.semanticscholar.org/search?year%5B0%5D=2012&year%5B1%5D=2023&q=information%20retrieval&sort=relevance'

# "https://www.semanticscholar.org/search?q=information%20retrieval&sort=relevance&yearFilter=2012%2F2022"
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, "html.parser")

# Find all the articles on the page
articles = soup.find_all("div", class_="result-page")
print(articles)

# Loop through each article and extract the relevant information
for article in articles:
    title = article.find("a", class_="title").text
    authors = article.find("div", class_="authors").text
    year = article.find("span", class_="year").text
    venue = article.find("span", class_="venue").text
    
    # Print the information for each article
    print("Title:", title)
    print("Authors:", authors)
    print("Year:", year)
    print("Venue:", venue)
    print("\n")
```
